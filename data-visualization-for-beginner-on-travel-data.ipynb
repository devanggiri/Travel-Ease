```python
# Load the libaries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the dataset
data = pd.read_csv("/kaggle/input/travel-dataset-guide-to-indias-must-see-places/Top Indian Places to Visit.csv")
```

```python
# Drop the first coloum
data = data.drop("Unnamed: 0",axis=1)
```

# Data Visualization

```python
plt.figure(figsize=(16,8))
sns.countplot(x="Significance",data=data)
plt.title("Most visited type of Place")
plt.xticks(rotation=90)
plt.show()
```

```python
data["Best Time to visit"].unique()
```

```python
data["Best Time to visit"]=data["Best Time to visit"].str.replace("All","Anytime")
data["Best Time to visit"]=data["Best Time to visit"].str.replace("Anytime ","Anytime")

```

```python
sns.countplot(x="Best Time to visit",data=data)
plt.title("Most Visited Time")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16,8))
Top_state = (data["State"].value_counts()).head(10)
Top_state.plot(kind="bar")
plt.title("Top State to visit")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16,8))
Top_city = (data["City"].value_counts()).head(10)
Top_city.plot(kind="bar")
plt.title("Top City to visit")
plt.xticks(rotation=90)
plt.show()
```
